```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
motor_df = pd.read_csv('data/NormalMotorOn_1.csv')
```

```python
encoder_data = motor_df['Volt1']
```

```python
plt.figure(figsize=(18, 8))
plt.axvline(x=640, color='r', linestyle='--')

plt.axvline(x=3130, color='r', linestyle='--')

plt.plot(encoder_data)
plt.xlim(200, 4000)
plt.show()
```

```python
square_data = [20 if x >= 2 else 0 for x in encoder_data]
```

```python
plt.figure(figsize=(18, 8))
plt.axvline(x=640, color='r', linestyle='--')

plt.axvline(x=3130, color='r', linestyle='--')


plt.plot(square_data)
plt.xlim(200, 4000)
plt.show()
```

```python
rfft = np.fft.rfft(encoder_data)
freqs = np.fft.rfftfreq(len(encoder_data), d=1/30000)
amps = np.abs(rfft)
```

```python
plt.figure(figsize=(18, 8))
plt.plot(freqs, amps)
plt.xlim(0, 500)
plt.ylim(0,4e8)
plt.show()
```
